```python
import pandas as pd
import os
import ast
```

```python
man_folder = 'C:/Users/oryan/Documents/esac-project/manifests'
pred_folder = 'C:/Users/oryan/Documents/esac-project/predictions'
```

```python
labels_df = pd.read_csv(f'{man_folder}/validation-set-manifest-checked.csv', index_col = 0)
predictions_df = pd.read_csv(f'{pred_folder}/validation-test.csv',index_col = 0).reset_index()
```

```python
pred_named = (
    predictions_df
    .assign(zooniverse_id = predictions_df.id_str.apply(lambda x: os.path.basename(x).replace('_300_300_3.png','')))
    .drop(columns='id_str')
)
```

```python
labels_df
```

```python
merged_df = (
    labels_df
    .merge(pred_named,on='zooniverse_id',how='left')
    .drop(columns=['RA', 'DEC','filepath'])
)
```

```python
binary_df = (
    merged_df
    .assign(binary_prediction = merged_df.interacting_pred.apply(lambda x: 1 if ast.literal_eval(x)[0] > 0.5 else 0))
)
```
