## Extracting GZ2 Mergers

```python
gz2_spec_only = pd.read_csv(f'{folder}/zoo2MainSpecz.csv')
```

```python
gz2_phot_only = pd.read_csv(f'{folder}/zoo2MainPhotoz.csv')
```

```python
gz2_full = pd.concat([gz2_phot_only,gz2_spec_only])
```

```python
gz2_red = gz2_full[['dr7objid','total_classifications','ra','dec','t08_odd_feature_a24_merger_weighted_fraction']]
```

```python
gz2_red.total_classifications.value_counts()
```

```python
gz2_reduced = (
    gz2_red
    .query('total_classifications >= 20')
    .query('t08_odd_feature_a24_merger_weighted_fraction > 0.95')
    .rename(columns={'dr7objid':'zooniverse_id','ra':'RA','dec':'Dec','t08_odd_feature_a24_merger_weighted_fraction':'interaction_votes'})
    .drop(columns='total_classifications')
)
```

```python
gz2_mergers = (
    gz2_reduced
    .assign(interaction = 1)
    .drop(columns = 'interaction_votes')
)
```

```python
gz2_mergers
```

```python
gzh_reduced = gzh_all[['zooniverse_id','RA','DEC', 'total_count', 't08_odd_feature_a06_merger_weighted_fraction']]
```

```python
gzc_mergers = (
    gzc_reduced
    .query('num_classifications > 20')
    .query('t16_merging_tidal_debris_a0_merging_weighted_frac > 0.95')
    .rename(columns={'ID' : 'zooniverse_id'})
    .drop(columns=['num_classifications', 't16_merging_tidal_debris_a0_merging_weighted_frac'])
    .assign(interaction = 1)
)
```

```python
gzc_mergers
```

## Galaxy Zoo DeCALS

```python
gzd_full = pd.read_csv(f'{folder}/gz_decals_volunteers_c.csv')
```

```python
gzd_full
```

```python
gzd_red = (
    gzd_full[['iauname','ra','dec','merging_total-votes','merging_none_fraction','merging_minor-disturbance_fraction','merging_major-disturbance_fraction','merging_merger_fraction']]
    .rename(columns={'merging_total-votes':'total_votes', 'merging_minor-disturbance_fraction':'minor_dist_frac','merging_major-disturbance_fraction':'major_dist_frac','merging_merger_fraction':'merging_frac'})
)
```

```python
gzd_mergers = (
    gzd_red
    .query('total_votes > 20')
    .query('merging_none_fraction <= 0.05')
)
```

```python
gzd_mergers
```

```python
gzd_mergers_check = gzd_mergers.assign(interaction_frac = gzd_mergers.apply(lambda row: row.minor_dist_frac + row.major_dist_frac + row.merging_frac, axis=1))
```

```python
gzd_final = (
    gzd_mergers_check
    .query('interaction_frac > 0.95')
    .rename(columns={'iauname' : 'zooniverse_id', 'ra': 'RA', 'dec' : 'Dec'})
    .drop(columns=['total_votes','merging_none_fraction','minor_dist_frac','major_dist_frac','merging_frac','interaction_frac'])
    .assign(interaction = 1)
)
```

```python
gzd_final
```

## Counting Contributions

```python
gz1_contr = len(gz1_mergers)
gz2_contr = len(gz2_mergers)
gzh_contr = len(gzh_queried)
gzc_contr = len(gzc_mergers)
gzd_contr = len(gzd_final)
total_gals = gz1_contr + gz2_contr + gzh_contr + gzc_contr + gzd_contr
```

```python
total_gals
```

## Combining All

```python
all_int_found = (
    pd.concat([gz1_mergers, gz2_mergers, gzh_queried, gzc_mergers, gzd_final])
    .reset_index()
    .drop(columns='index')
)
```

```python
all_int_found.to_csv(f'{folder}/all-interaction.csv')
```
