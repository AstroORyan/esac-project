```python
training_manifest = pd.read_csv(f'{folder}/gz-hubble-local-manifest.csv', index_col = 0)
full_manifest = pd.read_csv(f'C:/Users/oryan/Documents/zoobot/hubble-dataset-creation/data/gz_hubble_main.csv')
```

```python
full_manifest
```

```python
full_manifest_red = (
    full_manifest[['zooniverse_id','total_count','RA','DEC','t08_odd_feature_a06_merger_fraction','t08_odd_feature_a06_merger_weighted_fraction']]
    .rename(columns={'t08_odd_feature_a06_merger_fraction':'merger_fraction','t08_odd_feature_a06_merger_weighted_fraction':'weighted_merger_fraction'})
)
```

```python
full_manifest_red
```

```python
full_manifest_n = full_manifest_red.query('total_count > 100')
full_manifest_merg = full_manifest_n.query('weighted_merger_fraction >= 0.55')
full_manifest_no_merg = full_manifest_n.query('weighted_merger_fraction < 0.01')
```

```python
full_manifest_merg
```

```python
full_manifest_no_merg
```

```python
no_merg_sample_df = full_manifest_no_merg.sample(len(full_manifest_merg))
```

```python
validation_df = pd.concat([no_merg_sample_df, full_manifest_merg])
```

```python
validation_df
```

```python
binary_validation_set = validation_df.assign(interaction = validation_df.weighted_merger_fraction.apply(lambda x: 1 if x >= 0.55 else 0))
```

```python
binary_validation_set.interaction.value_counts()
```

```python
export_df = shuffle(binary_validation_set.drop(columns=['total_count','merger_fraction','weighted_merger_fraction'])).reset_index().drop(columns='index')
```

```python
export_df
```

```python
export_df.to_csv(f'{folder}/large-training-set.csv')
```
