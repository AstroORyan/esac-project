```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\oryan\Documents\zoobot_new\manifests\gz-mergers-coords-all-62.csv',index_col=0)
```

```python
df
```

```python
df_prim = (
    df[['Names','Prim_RA','Prim_DEC']]
    .rename(columns={'Prim_RA' : 'RA', 'Prim_DEC':'Dec'})
    .assign(Names_tmp = df.Names.apply(lambda x: x.strip() + '_Prim'))
    .drop(columns='Names')
    .rename(columns={'Names_tmp' : 'Names'})
)
```

```python
df_prim
```

```python
df_sec = (
    df[['Names','Sec_RA','Sec_DEC']]
    .rename(columns={'Sec_RA' : 'RA', 'Sec_DEC':'Dec'})
    .dropna()
    .assign(Names_tmp = df.Names.apply(lambda x: x.strip() + '_Sec'))
    .drop(columns='Names')
    .rename(columns={'Names_tmp' : 'Names'})
)
```

```python
df_sec
```

```python
df_export = pd.concat([df_prim, df_sec])
```

```python
df_export
```

```python
df_export.to_csv('C:/Users/oryan/Documents/zoobot_new/manifests/gz-mergers-coords-all-seperated.csv')
```
